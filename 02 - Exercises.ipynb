### Q2

Create 2 numpy arrays named `M4` and `M5`.

* `M4` should be a 3x5 matrix of *uniformly distributed* random numbers ranging from 5 to 15.

* `M5` should be a 4x3 matrix of *normally distributed* random numbers with mean 20 and 7 standard deviation.

Using the above, write code that does the following:

* calculate the sum total of `M4` *(the answer should be 147.6)*.

* find out how many numbers in `M5` are greater than 21 *(the answer should be 2)*.

* do a matrix multiplication of `M5` and `M4`, and calculate the mean *(the answer should be 437.9)*.

```python
rng = np.random.default_rng(50)
# add your code below
M4 = rng.uniform(5, 15, size = [3,5])
M5 = rng.normal(loc = 20, scale = 7, size=[4,3])
print(np.sum(M4))
print(np.sum(M5>=21))
print(M5 @ M4)
print(np.round(np.mean(M5 @ M4),1))
```

Use Matplotlib to plot a cosine wave from 0 to 2Ï€.

* The plot should be a dotted line and in red colour.

* Add gridlines only for the y-axis (horizontal)

```python
import matplotlib.pyplot as plt
# add your code below
x = np.linspace(0, 2*np.pi, 1000)
plt.plot(x, np.cos(x), linestyle = 'dotted', color="red")
plt.grid(axis ='y')
```
